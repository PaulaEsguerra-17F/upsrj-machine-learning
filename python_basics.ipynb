# Importacion de librerias
Este bloque importa las librerias numpy, pandas y time.

```python
import numpy as np
import pandas as pd
import time
```

# MEDICIÓN DE TIEMPO DE EJECUCIÓN
Se está midiendo el tiempo exacto que le toma a tu computadora procesar una lista de 100 millones de números realizando 200 millones de sumas.

```python
lista = list(range(100000000))

inicio = time.perf_counter()

suma = 0
for i in lista:
  suma += i
  suma += 1

final = time.perf_counter()

tiempo = final - inicio

print(f"tiempo de ejecucuin: {tiempo:.6f} sec")
```

# OPTIMIZACIÓN MEDIANTE VECTORIZACIÓN
Se mide el rendimiento de NumPy procesando 1,000 millones de datos (10 veces más que el anterior). Se utiliza vectorización para realizar la suma de forma masiva en el procesador, demostrando cómo una librería especializada es drásticamente más rápida que un bucle for tradicional.

```python
np_lista = np.arange(1000000000)

inicio = time.perf_counter()

suma_np = np.sum(np_lista)

final = time.perf_counter()

tiempo = final - inicio

print(f"tiempo de ejecucuin: {tiempo:.6f} sec")
```

# SIMULACIÓN DE BASE DE DATOS E INDEXACIÓN MANUAL
Se organiza la información en una matriz (lista de listas) y se utilizan diccionarios como índices. Esto permite consultar datos de forma legible (usando nombres de filas y columnas) en lugar de usar números de posición, simulando el funcionamiento básico de una tabla de base de datos o un DataFrame.

```python
base_de_datos = [
["Oscar", 21, "Gatos"],
["Paula", 21, "Tiburones"],
["Jonatan", 24, "Hurones"],
["Kevin", 22, "Periquitos"],
["Lucero", 21, "Perros"],
["Israel", 21, "Lobos"],
["Sergio", 21, "Perros"],
["Vahir", 21, "Leones"],
["Diego", 21, "Pinguinos"]
]

columnas = {"Nombre":0, "Edad":1, "Animal":2}
filas    = {"Oscar":0, "Paula":1, "Jonatan":2, "Kevin":3, "Lucero":4, "Israel":5, "Sergio":6, "Vahir":7, "Diego":8}
print(base_de_datos[filas["Kevin"]][columnas["Animal"]])
```

# CONSULTA INTELIGENTE CON PANDAS
Convertimos los datos en una tabla automática. Esto nos permite buscar información directamente por su nombre (por ejemplo, saber qué animal tiene 'Kevin') sin tener que saber en qué posición exacta está.

```python
df = pd.DataFrame(base_de_datos, columns=["Nombre", "Edad", "Animal"], index=range(1,10))


print(df.Animal[df['Nombre'] == "Kevin"])
```
