# Importacion de librerias
Se usan las librerías para crear, entrenar y evaluar un modelo de Regresión Lineal con visualización de datos.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

Se crean los datos de entrada (X) y de salida (Y) para una regresión lineal.

X se convierte en un arreglo de dos dimensiones porque así lo requiere Scikit-Learn, y Y contiene los valores correspondientes a cada dato de X.

```python
# y = m * x + b

# Variable independiente
# Scikit-Learn requiere que se usen arreglos de 2D
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
print(X)

# Variable dependiente
Y = np.array([2, 4, 5, 4, 5])
print(Y)
```

Se grafica la relación entre X y Y usando un diagrama de dispersión.
Los puntos representan los datos originales.

```python
plt.scatter(X, Y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Datos originales")
plt.grid(True)
plt.show()
```

Se crea un modelo de regresión lineal con los datos X y Y.

Después, se obtienen los parámetros del modelo: la pendiente (m) y la intersección (b), que definen la ecuación de la recta ajustada a los datos.

```python
# Generamos nuestro modelo de regresion linear
model = LinearRegression()

# Alimentamos nuestro modelo
model.fit(X, Y)

# Parametros: m (peso), b (sesgo)
m = model.coef_[0]
b = model.intercept_

print(f"Pendiente (m) = {m: .2f}")
print(f"Independiente (b) = {b: .2f}")
```

Se generan las predicciones del modelo usando los valores de X y luego se evalúa su desempeño.

Se calcula el error cuadrático medio (MSE) para medir el error de las predicciones y el coeficiente R2 para saber qué tan bien el modelo se ajusta a los datos.

```python
# Generamos una prediccion
y_pred = model.predict(X)

# Evaluamos esa prediccion
mse = mean_squared_error(Y, y_pred)
r2 = r2_score(Y, y_pred)

print(f"Mean Squared Error = {mse: .2f}")
print(f"R2 Score = {r2: .2f}")
```

Se muestran los datos reales con puntos y se dibuja la recta del modelo de regresión lineal en color rojo.

La gráfica permite comparar visualmente los valores reales con las predicciones del modelo.

```python
plt.scatter(X, Y, label="Datos reales")
plt.plot(X, y_pred, color="red", label="Modelo lineal")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Regresion linear simple")
plt.grid(True)
plt.show()
```
