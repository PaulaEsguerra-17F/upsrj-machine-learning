```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# 1. definiremos las variables dependientes e independientes de nuestro modelo
# variable independiente X: distancia

X = np.arange(1, 21).reshape(-1, 1)

# variable dependiente y: precio miles de $
y = np.array([ 320, 310, 295, 285, 270, 260, 250, 240, 230, 225, 215, 210, 205, 200, 195, 190, 185, 180, 175, 170])
print(y)
```

```python
#generamos una predicción
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
#Evaluamos esa predicción
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y,y_pred)
print(f"Mean Squared Error:{mse:2f}")
print(f"R2 Score:{r2:2f}")
```

```python
plt.scatter(X,y,label="Datos reales")
plt.plot(X,y_pred,color="red",label="Modelo lineal")

#
plt.xlabel("X")
plt.ylabel("y")
plt.title("Regresion Lineal simple")
plt.grid(True)
plt.show()
```

# Ejercicio 2: Limite de la regresion lineal simple.

Ahora analizaremos como afecta la distancia al centro al precio de una casa en una ciudad donde ocurre lo siguiente:

-Si estan muy cerca del centro los precios son altos.

-Si esta en una zona intermedia los precios son mas bajos.

-Si esta muy lejos del centro los precios vuelven a ser altos, dado que se trata de residenciales exclusivos.

```python
# 1. definiremos las variables dependientes e independientes de nuestro modelo
# variable independiente X: distancia

X = np.arange(1, 21).reshape(-1, 1)
print(X)

# variable dependiente y: precio miles de $
y = np.array([340, 320, 290, 260, 230,
              210, 220, 245, 275, 305,
              330, 350, 365, 375, 380,
              385, 390, 395, 398, 400])
print(y)
```

```python
#Generamos nuestro modelo de regresion linear
model= LinearRegression()
#alimentamos nuestro modelo

model.fit(X,y)

#parametros:m(peso), b(sesgo)

m=model.coef_[0]
b = model.intercept_

print(f"Pendiente (m):{m:2f}")
print(f"Intercepto(b):{b:2f}")

#Graficamos:
plt.scatter(X,y)
plt.xlabel("Distancia al centro (km)")
plt.ylabel("Costo miles de $")
plt.title("Datos")
plt.grid(True)
plt.show()
```

# Implementación de Pipeline

Un pipeline encapsula todo el flujo de procesamiento de datos en un solo objeto.

Para un usuario, entrenar un pipeline es igual que entrenar un modelo.

El escalado no cambia el resultado conceptual, pero prepara el terreno para modelos mas complejos.

## Justificación

Hasta ahora hemos entrenado modelos paso por paso.
En la vida real, esos pasos siempre se repiten en el mismo orden. Un pipeline es una forma de decirle a la computadora: **Estos son los pasos, siempre ejecutalos asi**

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```
